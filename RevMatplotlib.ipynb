```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#generating data for x and y axis so that we can plot data and how it trends
x=np.linspace(0,10 ,300) #0 SE LEKAR 10 TAK 200 DATA GENERATE KAR DENGE
x
```

```python
y=np.sin(x)
y
```

```python
#its very hard to analyze according to the data instead we can use graphs to analyse it
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("this is my title")
plt.show()
```

```python
#kuch random data ko generate karte hein
x=np.random.rand(65)
y=np.random.rand(65)
plt.scatter(x,y)
```

```python
#agar random data ko color dena hai
colours=np.random.rand(65)
plt.scatter(x,y,c=colours) #c basically colors dene ke lie use hota hai
```

```python
#sizes ko bhi control kia ja sakta hai by using s
colours=np.random.rand(65)
sizes=np.random.rand(65)
plt.scatter(x,y,c=colours,s=sizes)
```

```python
#agar hame bada size chahiye to hum ky kar sakte hein ki
colours=np.random.rand(65)
sizes=1000*np.random.rand(65)
plt.scatter(x,y,c=colours,s=sizes)
```

```python
#there is one more parameter called alpha that represents intensity
colours=np.random.rand(65)
sizes=1000*np.random.rand(65)
plt.scatter(x,y,c=colours,s=sizes,alpha=0.5)
plt.xlabel("x axis")
plt.ylabel("y axis")
```
